```python
columns=np.arange(1990,2019).tolist()
columns_str = map(str,columns)
list_of_year = list(columns_str)
```

```python
lista =[]
for i in list_of_year:
    lista.append(df[i].isna().sum())
lista
### We can see that some years have countries with many missing values
```

```python
df1 = df.dropna(axis=0, how="any", thresh=25, subset=None, inplace=False)
### We drop rows with many Nan values
### Consider that this might not be necessary due to the fact that very few countries emmit the most amount of 
### CO2, so you can keep the Nan values and proceed to the next steps
df1.shape
```

```python
co2_region_df = df.groupby('Region')[list_of_year].mean()
```

```python
pd.set_option('display.max_columns', 30)
```

```python
co2_region_df
### Interesting how een in crisis the CO2 didn't dwindle.
```

```python
plt.figure(figsize=(15,15))
plt.rcParams['font.size'] = 10
plt.title("Mean CO2 emissions by region",fontweight="bold",fontdict={'fontsize': 30})
sns.heatmap(co2_region_df,linecolor='white',linewidths=1,annot=True)
plt.xlabel('Years',fontdict={'fontsize': 25})
plt.ylabel('Regions',fontdict={'fontsize': 25})
```

```python
plt.figure(figsize=(12,5))
plt.title('Countries with highest CO2 emissions (metric tons per capita)')
plt.ylabel('CO2 emissions (metric tons per capita) in 2019')
df.groupby('Country Name')['2019'].mean().sort_values(ascending=False)[0:15].plot(kind='line', marker='*', color='red', ms=10)
df.groupby('Country Name')['2019'].mean().sort_values(ascending=False)[0:15].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.title('Countries with highest CO2 emissions (metric tons per capita)')
plt.ylabel('CO2 emissions (metric tons per capita) in 2015')
df.groupby('Country Name')['2015'].sum().sort_values(ascending=False)[0:15].plot(kind='line', marker='*', color='red', ms=10)
df.groupby('Country Name')['2015'].sum().sort_values(ascending=False)[0:15].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.title('Countries with highest CO2 emissions (metric tons per capita)')
plt.ylabel('CO2 emissions (metric tons per capita) in 2005')
df.groupby('Country Name')['2005'].mean().sort_values(ascending=False)[0:15].plot(kind='line', marker='*', color='red', ms=10)
df.groupby('Country Name')['2005'].mean().sort_values(ascending=False)[0:15].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.title('Countries with highest CO2 emissions (metric tons per capita)')
plt.ylabel('CO2 emissions (metric tons per capita) in 1995')
df.groupby('Country Name')['1995'].mean().sort_values(ascending=False)[0:15].plot(kind='line', marker='*', color='red', ms=10)
df.groupby('Country Name')['1995'].mean().sort_values(ascending=False)[0:15].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```
